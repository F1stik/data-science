```python
# Одна запись о продажах
sale_record = {
    "date": "2026-07-01",
    "product": "Ноутбук",
    "quantity": 3,
    "price": 75000,
    "region": "Москва",
    "manager": "Витаутас"
}

print(sale_record["product"])
print(sale_record["quantity"] * sale_record["price"])
print(sale_record["price"], sale_record['region'])
```

```python
sales = [
    {"date": "2026-07-01", "product": "Ноутбук", "quantity": 3, "price": 75000},
    {"date": "2026-07-02", "product": "Мышь", "quantity": 15, "price": 1200},
    {"date": "2026-07-03", "product": "Ноутбук", "quantity": 2, "price": 75000},
    {"date": "2026-07-04", "product": "Клавиатура", "quantity": 8, "price": 4500}
]
##Всё продукты
for sale in sales:
    print(sale['product'])
##Выручка
total_revenue = 0
for sale in sales:
    revenue = sale['quantity'] * sale['price'] 
    total_revenue += revenue
print(f'Общая выручка составила {total_revenue}')
##С максимальной выручкой
max_zen = 0
best_sale = None
for sale in sales:
    revenue = sale['quantity'] * sale['price']
    if revenue > max_zen:
        max_zen = revenue
        best_sale = sale
print('максимальная выручка', max_zen)
print('лучшая продажа:', best_sale)
```

```python
def total_revenue(sales_list):
    total = 0
    for sale in sales_list:
        total += sale["quantity"] * sale["price"]
    return total

print(total_revenue(sales))
```
