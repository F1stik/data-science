```python
# Задание 4
def calculate_stats(data):
    """Возвращает базовую статистику по списку чисел"""
    
    total = sum(data)                    # сумма всех значений
    count = len(data)                    # количество элементов
    average = round(total / count, 2)    # среднее (округляем до 2 знаков)
    maximum = max(data)                  # максимальное значение
    minimum = min(data)                  # минимальное значение
    
    # Считаем медиану вручную (без дополнительных библиотек)
    sorted_data = sorted(data)           # сортируем список
    n = len(sorted_data)
    if n % 2 == 1:                       # если нечётное количество
        median_value = sorted_data[n // 2]
    else:                                # если чётное количество
        median_value = (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2
        median_value = round(median_value, 2)
    
    return {
        "sum": total,
        "count": count,
        "average": average,
        "max": maximum,
        "min": minimum,
        "median": median_value
    }


# Вызываем функцию
stats = calculate_stats(monthly_sales)
print(stats)
```
