```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r'D:\Python\project\data-science\data\raw\SampleSuperstore.csv')
```

```python
plt.figure(figsize= (10, 5))
sns.barplot(data=df,
            x='Category',
            y='Profit', 
            estimator='sum',
            hue='Category',
            palette='Set2'
            )
plt.title('Общая прибыль по категориям')
plt.legend(title= 'Категория')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(data=df,
            x = 'Category',
            y='Profit',
            estimator='sum',
            hue='Region',
            palette='Set1'
              )
plt.title('Общая прибыль по категориям и регионам')
plt.legend(title='Регион', bbox_to_anchor = (1.05,1))
plt.show
```

```python
plt.figure(figsize=(11, 6))
sns.barplot(
    data=df, 
    x='Category', 
    y='Profit', 
    estimator='sum',
    hue='Region',
    palette='Set2'
)
plt.title('Общая прибыль по категориям и регионам', fontsize = 14)
plt.xlabel('Категория товаров')
plt.ylabel('Прибыль ($)')
plt.legend(title = 'Регион',bbox_to_anchor = (1.02,1),loc='upper left')
plt.xticks(rotation = 15)
plt.tight_layout()
plt.show
```

```python
plt.figure(figsize=(11, 6))
sns.barplot(
    data=df, 
    x='Category', 
    y='Profit', 
    estimator='sum',
    hue='Region',
    palette='Set2'
)
plt.title('Общая прибыль по категориям и регионам', fontsize = 14)
plt.xlabel('Категория товаров')
plt.ylabel('Прибыль ($)')
plt.legend(title = 'Регион',bbox_to_anchor = (1.02,1),loc='upper left')
plt.xticks(rotation = 15)
plt.axhline(y=0,color = 'red',linestyle = '--',linewidth=1.5)
plt.tight_layout()
plt.show
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(
    data=df, 
    x='Discount', 
    y='Profit', 
    hue='Category',
    alpha=0.6
)
plt.title('Связь между скидкой и прыбылью')
plt.axhline(y=0,color='red',linestyle='--')
plt.xlabel('Скидка')
plt.ylabel('Прибыль ($)')
plt.legend(title = 'Категория')
plt.show
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(
    data=df, 
    x='Discount', 
    y='Profit', 
    hue='Category',
    size='Quantity',
    alpha=0.5,
    sizes=(20,200)
    
)
plt.title('Связь между скидкой и прыбылью')
plt.axhline(y=0,color='red',linestyle='--')
plt.xlabel('Скидка')
plt.ylabel('Прибыль ($)')
plt.legend(title = 'Категория')
plt.show
```

```python
numeric_df = df[['Sales','Quantity','Discount','Profit']]
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(),annot = True, cmap='coolwarm',center=0)
plt.title('Корееляция между числовыми показателями')
plt.show
```

```python
numeric_df = df[['Sales', 'Quantity', 'Discount', 'Profit']]

plt.figure(figsize=(8, 6))
sns.heatmap(
    numeric_df.corr(), 
    annot=True,           # показывать числа
    cmap='coolwarm',      # цветовая схема
    center=0,             # центр цветовой шкалы
    fmt=".2f"             # два знака после запятой
)
plt.title('Корреляция между числовыми показателями продаж')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Category', y='Profit')
plt.title('Распределение прибыли по категориям')
plt.axhline(y=0, color='red', linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Медианная прибыль
sns.boxplot(data=df, x='Category', y='Profit', showfliers=False)

# Добавляем среднее значение точками
sns.pointplot(data=df, x='Category', y='Profit', 
              estimator='mean', color='red', markers='D')

plt.title('Медианная и средняя прибыль по категориям')
plt.axhline(y=0, color='red', linestyle='--')
plt.show()
```

```python
df['Order Month'] = pd.to_datetime(df['Order Date']).dt.to_period('M')
plt.figure(figsize=(12,8))
sns.lineplot(data=df,x='Order Month',y='Profit',estimator='sum')
plt.title('Динамика общей прибыли по месяцам')
plt.xticks(rotation=45)
plt.show
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Category', palette='Set2')
plt.title('Количество заказов по категориям')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df, 
    x='Discount', 
    y='Profit', 
    hue='Category',
    size='Quantity',      # размер точки = количество товаров
    alpha=0.5,
    sizes=(20, 200)
)
plt.title('Связь между скидкой и прибылью (размер точки = кол-во)')
plt.axhline(y=0, color='red', linestyle='--', linewidth=1.5)
plt.xlabel('Скидка')
plt.ylabel('Прибыль ($)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize=(14,5))
#Гарфик №1
sns.barplot(data = df,
            x='Category',
            y='Profit',
            estimator='sum', 
            ax=axes[0],
            palette='Set2'
            )
axes[0].set_title('Общая прибыль')
axes[0].axhline(0,color='red',linestyle='--')
#График №2
sns.countplot(data=df,
              x='Category',
              ax=axes[1],
              palette='Set2'
              )
axes[1].set_title('Количество заказов по категориям')
plt.tight_layout()
plt.show
```
