```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r'D:\Python\project\data-science\data\raw\SampleSuperstore.csv')
```

```python
plt.figure(figsize= (10, 5))
sns.barplot(data=df,
            x='Category',
            y='Profit', 
            estimator='sum',
            hue='Category',
            palette='Set2'
            )
plt.title('Общая прибыль по категориям')
plt.legend(title= 'Категория')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(data=df,
            x = 'Category',
            y='Profit',
            estimator='sum',
            hue='Region',
            palette='Set1'
              )
plt.title('Общая прибыль по категориям и регионам')
plt.legend(title='Регион', bbox_to_anchor = (1.05,1))
plt.show
```

```python
plt.figure(figsize=(11, 6))
sns.barplot(
    data=df, 
    x='Category', 
    y='Profit', 
    estimator='sum',
    hue='Region',
    palette='Set2'
)
plt.title('Общая прибыль по категориям и регионам', fontsize = 14)
plt.xlabel('Категория товаров')
plt.ylabel('Прибыль ($)')
plt.legend(title = 'Регион',bbox_to_anchor = (1.02,1),loc='upper left')
plt.xticks(rotation = 15)
plt.tight_layout()
plt.show
```

```python
plt.figure(figsize=(11, 6))
sns.barplot(
    data=df, 
    x='Category', 
    y='Profit', 
    estimator='sum',
    hue='Region',
    palette='Set2'
)
plt.title('Общая прибыль по категориям и регионам', fontsize = 14)
plt.xlabel('Категория товаров')
plt.ylabel('Прибыль ($)')
plt.legend(title = 'Регион',bbox_to_anchor = (1.02,1),loc='upper left')
plt.xticks(rotation = 15)
plt.axhline(y=0,color = 'red',linestyle = '--',linewidth=1.5)
plt.tight_layout()
plt.show
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(
    data=df, 
    x='Discount', 
    y='Profit', 
    hue='Category',
    alpha=0.6
)
plt.title('Связь между скидкой и прыбылью')
plt.axhline(y=0,color='red',linestyle='--')
plt.xlabel('Скидка')
plt.ylabel('Прибыль ($)')
plt.legend(title = 'Категория')
plt.show
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(
    data=df, 
    x='Discount', 
    y='Profit', 
    hue='Category',
    size='Quantity',
    alpha=0.5,
    sizes=(20,200)
    
)
plt.title('Связь между скидкой и прыбылью')
plt.axhline(y=0,color='red',linestyle='--')
plt.xlabel('Скидка')
plt.ylabel('Прибыль ($)')
plt.legend(title = 'Категория')
plt.show
```
