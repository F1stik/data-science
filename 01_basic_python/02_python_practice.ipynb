```python
import pandas as pd

# Таблица 1 — продажи
sales = pd.DataFrame({
    'order_id': [101, 102, 103, 104],
    'product': ['A', 'B', 'A', 'C'],
    'amount': [1500, 2300, 1800, 900]
})

# Таблица 2 — информация о товарах
products = pd.DataFrame({
    'product': ['A', 'B', 'C', 'D'],
    'category': ['Электроника', 'Одежда', 'Электроника', 'Дом'],
    'price': [1500, 2300, 1800, 1200]
})

print("Продажи:")
print(sales)
print("\nТовары:")
print(products)
```

```python
merged = pd.merge(
    sales,
    products,
    on='product',
    how='left'
)

print(merged)
```

```python
# 1. left — оставляем все продажи
merged_left = pd.merge(sales, products, on='product', how='left')
print("how='left':")
print(merged_left)
print()

# 2. inner — только совпадающие товары
merged_inner = pd.merge(sales, products, on='product', how='inner')
print("how='inner':")
print(merged_inner)
```

```python
# Добавляем новый заказ с товаром E (его нет в products)
sales2 = pd.concat([
    sales,
    pd.DataFrame({'order_id': [105], 'product': ['E'], 'amount': [500]})
], ignore_index=True)

print("Новые продажи:")
print(sales2)
print()

# left
print("how='left':")
print(pd.merge(sales2, products, on='product', how='left'))
print()

# inner
print("how='inner':")
print(pd.merge(sales2, products, on='product', how='inner'))
```

```python
print("how='outer':")
print(pd.merge(sales2, products, on='product', how='outer'))
```

```python
# Таблица продаж (ключ называется product_code)
sales3 = pd.DataFrame({
    'order_id': [101, 102, 103],
    'product_code': ['A', 'B', 'C'],
    'amount': [1500, 2300, 900]
})

# Таблица товаров (ключ называется product)
products3 = pd.DataFrame({
    'product': ['A', 'B', 'C'],
    'category': ['Электроника', 'Одежда', 'Электроника']
})

# Объединяем
merged3 = pd.merge(
    sales3,
    products3,
    left_on='product_code',
    right_on='product',
    how='left'
)

print(merged3)
```

```python
merged3 = pd.merge(
    sales3,
    products3,
    left_on='product_code',
    right_on='product',
    how='left'
)

# Удаляем дублирующий столбец
merged3 = merged3.drop(columns=['product'])

print(merged3)
```

```python
# Две таблицы одинаковой структуры
sales_jan = pd.DataFrame({
    'order_id': [101, 102],
    'amount': [1500, 2300]
})

sales_feb = pd.DataFrame({
    'order_id': [103, 104],
    'amount': [1800, 900]
})

# Склеиваем по строкам
all_sales = pd.concat([sales_jan, sales_feb], ignore_index=True)

print(all_sales)
```

```python
all_sales_old_index = pd.concat([sales_jan, sales_feb])
print(all_sales_old_index)
```
