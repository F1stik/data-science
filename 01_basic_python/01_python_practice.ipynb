```python
sales = [
    {"date": "2026-07-01", "product": "Ноутбук", "quantity": 3, "price": 75000},
    {"date": "2026-07-02", "product": "Мышь", "quantity": 15, "price": 1200},
    {"date": "2026-07-03", "product": "Ноутбук", "quantity": 2, "price": 75000},
    {"date": "2026-07-04", "product": "Клавиатура", "quantity": 8, "price": 4500}
]
##Всё продукты
for sale in sales:
    print(sale['product'])
##Выручка
total_revenue = 0
for sale in sales:
    revenue = sale['quantity'] * sale['price'] 
    total_revenue += revenue
print(f'Общая выручка составила {total_revenue}')
##С максимальной выручкой
max_zen = 0
best_sale = None
for sale in sales:
    revenue = sale['quantity'] * sale['price']
    if revenue > max_zen:
        max_zen = revenue
        best_sale = sale
print('максимальная выручка', max_zen)
print('лучшая продажа:', best_sale)
```

```python
def get_product_names(data):
    """Возвращает список названий всех продуктов"""
    names = []
    for sale in data:
        names.append(sale["product"])
    return names


result = get_product_names(sales)
print(result)
```

```python
def analyze_sales(data):
    total_revenue = sum(sale['quantity'] * sale['price'] for sale in data)   
    unique_products = len(set(sale["product"] for sale in data))
    most_expensive = max(data, key=lambda x: x["price"])["product"]
    
    return {
    "total_revenue": total_revenue,
    "unique_products": unique_products,      # количество уникальных товаров
    "most_expensive": most_expensive        # название самого дорогого товара
    }
result_two = analyze_sales(sales)
print(result_two)
```
