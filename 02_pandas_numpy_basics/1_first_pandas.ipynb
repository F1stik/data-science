Первый код: создаём DataFrame в pandas

```python
import pandas as pd
data = {
    'order_id': [1001, 1002, 1003, 1004, 1005],
    'datetime': ['2025-03-05 08:15', '2025-03-05 09:40', 
                 '2025-03-05 14:30', '2025-03-05 07:55', '2025-03-06 10:20'],
    'product': ['Капучино', 'Латте', 'Эспрессо', 'Капучино', 'Латте'],
    'price': [180, 210, 120, 180, 210],
    'quantity': [2, 1, 1, 1, 3],
    'total': [360, 210, 120, 180, 630]
}
```

```python
df = pd.DataFrame(data)
```

```python
print('первые строки данных')
print(df.head())
```

```python
print("\n Информация о данных")
print(df.info())
```

```python
print(df.describe())
```

```python
import pandas as pd
from pathlib import Path

# Путь к файлу относительно ноутбука (из notebooks/ в data/raw/)
csv_path = Path("../data/raw/coffee_shop_orders.csv")

df = pd.read_csv(csv_path)

print("Данные успешно загружены!")
print(f"Размер датасета: {df.shape[0]} строк и {df.shape[1]} колонок")
print("\nПервые 5 строк:")
print(df.head())
```

```python
print(df['product'].value_counts())
```

```python
df['hour'] = pd.to_datetime(df['datetime']).dt.hour
print(df['hour'].value_counts().sort_index())
```
