Практика с реальный data set с сайта kaggle "Sample Superstore Dataset"
 **Цель**: Подготовить данные к обратобке и постараться сделать базовые вычисления для анализа данных


```python
import pandas as pd
```

```python
df = pd.read_csv(r'D:\Python\project\data-science\data\raw\SampleSuperstore.csv')
```

```python
print(df.head())
```

```python
print(df.shape)
```

```python
print(df[['Sales', 'Profit' ]].describe())
```

```python
print(df.info)
```

```python
print(df.isnull().sum())
```

```python
print('Выручка по регионам')
print(df.groupby('Region')['Profit'].sum())
print('Продажи по регионам')
print(df.groupby('Region')['Sales'].sum())
print('Среднии Продажи по регионам')
print(df.groupby('Region')['Sales'].mean())
```

```python
print('Средняя скидка по регионам')
print(df.groupby('Region')['Discount'].mean())
```

```python
print('Средняя прибль при разных уровнях скидки')
print(df.groupby('Discount')['Profit'].mean().sort_values())
```
